# **Ejercicio N° 1:**

```python
import pandas as pd #importamos pandas para usar dataframe
import numpy as np #importamos numpy para usar matrices
from datetime import datetime, timedelta

# Establecer la semilla para reproducibilidad
np.random.seed(0)

# con esto establecememos la variable "hoy" con la fecha de  hoy
hoy = datetime.today()

#con estos estableces una serie con los ultimos 5 dias a partir de hoy la cual la usaremos para los indices del dataframe
fechas = [hoy - timedelta(days=i) for i in range(5)]

fechas.reverse()  # Ordenamos los dias cronlogicamente

# Convertimos las fechas en strig
fechas_str = [fecha.strftime("%Y-%m-%d") for fecha in fechas]

# Crear un DataFrame con datos aleatorios
df1 = pd.DataFrame(
    data=np.random.randint(1, 20, (5,1 )),  # valores entre 0 y 20, en 5 filas y 1 columna
    columns=["NumeroUnid"],  # agregamos (temporalmente solo una columna llamada "NumeroUnid"
    index=fechas_str
)

# agregamos una columna llamada "Produc" con los valores:"Bolsos","Zapatos","Camisas","Enaguas","Sombreros"]
df1["Produc"] = ["Bolsos", "Zapatos", "Camisas", "Enaguas", "Sombreros"]

# agregamos una columna llamada PrecioUni la cual seran valores aleatorios que van del 1 al 100 redondeados en 2 decimales
df1["PrecioUni"] = np.round(np.random.uniform(1, 100, size=5), 2)

# agregamos una columna llamada TotalSinDesc la cual es una multiplicacion de la columna NumeroUnid por PrecioUni
df1["TotalSinDesc"] = df1["NumeroUnid"] * df1["PrecioUni"]

# agregamos una columna llamada TotalConDesc para mostras el precio ya rebajando el 5%
df1["TotalConDesc"] = np.round(df1["TotalSinDesc"] * 0.95, 2)

# Reordenamos las columnas para que se vean mejor
columnas = ["Produc", "NumeroUnid", "PrecioUni", "TotalSinDesc", "TotalConDesc"]
df1 = df1[columnas]


# Mostrar el DataFrame
df1
```

# **Ejercicio N° 2**

```python
import pandas as pd #importamos pandas para crear series

# Diccionario con los Supuestos Ingresos
ingresos = {
    "Bill Gates": 10000, "Mark Twain": 8000, "Elon Musk": 9500, "Ada Lovelace": 7500, "Marie Curie": 8200,
    "Leonardo da Vinci": 8700, "Steve Jobs": 9200, "Albert Einstein": 8800, "Grace Hopper": 7800, "Nikola Tesla": 9100,
    "Jane Austen": 7200, "Isaac Newton": 9300, "Coco Chanel": 8400, "Michael Jordan": 8900, "Amelia Earhart": 8600
}

# Crear la serie Singressos partir del diccionario de ingresos
serie_ingresos = pd.Series(ingresos)

# Calculamos el ingreso mínimo, máximo y media
minimo = serie_ingresos.min()
maximo = serie_ingresos.max()
media = serie_ingresos.mean()

# Creamos otra serie con los datos estadísticos
estadisticas = pd.Series({'Mínimo': minimo, 'Máximo': maximo, 'Media': media})

# Mostramos un encabezado
print("Estadisticas de Ingresos:")

# Mostramos la serie "estadisticas"con los datos redondeados
print(np.round(estadisticas),2);
```

# **Ejercicio N° 3**

```python
import pandas as pd #importamos pandas para usar dataframe

# Crear el diccionario con los datos
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [30500, 35600, 28300, 33900, 42500],
    'Gastos': [22000, 23450, 18100, 35700, 32450]
}

# Crear el DataFrame en base al diccionario "data"
df = pd.DataFrame(data)

#agregamos la columna 1 Gasto/Ventas*100
df["Columna 1"] = df["Gastos"] / df["Ventas"] * 100

#agregamos la columna 2 Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
df['Columna 2'] = ['Meta Superada' if ventas >= 30000 else 'Meta no Superada' for ventas in df['Ventas']]

#agregamos la columna 2 Si Gasto>Ventas, calcular el sobre gasto.
df['Columna 3'] = df.apply(lambda row: row['Gastos'] - row['Ventas'] if row['Gastos'] > row['Ventas'] else 0, axis=1)

# Ventas de los ultimos 3 meses
ventas_ultimos_3_meses = df.loc[2:4, 'Ventas'].sum()

# Gastos de febrero a abril
total_gastos_feb_abr = df.loc[1:3, 'Gastos'].sum()

# Mostramos los totales
print("Total de ventas de los últimos 3 meses:", ventas_ultimos_3_meses)
print("Total de gastos de febrero a abril:", total_gastos_feb_abr)
print(" ")

print("DataFrame con nuevas columnas:")

df
```
